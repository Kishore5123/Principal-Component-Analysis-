```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv("wine.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
data['Type'].value_counts()
```

```python
data1=data.iloc[:,1:]
```

```python
data1.head()
```

```python
data1.shape
```

```python
data1.info()
```

```python
#data convering into numpy array
data1_arr=data1.values
```

```python
data1_arr
```

```python
#Normalizing the numerical data
data1_norm=scale(data1_arr)
data1_norm
```

# PCA Implementation 

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

data1_pca=pca.fit_transform(data1_norm)
data1_pca
```

```python
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color='magenta')
plt.show()
```

```python
# Final Dataframe
final_df=pd.concat([data['Type'],pd.DataFrame(data1_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# Checking with Other Clustering Algorithms 

# 1.Hierarchical 

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(data1_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(data1_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
data2=data.copy()
data2['clustersid']=hclusters.labels_
data2
```

# 2.K_means Clustering

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data1_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(data1_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
data4=data.copy()
data4['clusters3id']=clusters3.labels_
data4
```
